```python
#Riemann integral Method

import numpy as np

a = 0.2
b = 5
n = 10
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = (x-1)**2/np.exp(x)

I_riemannL = h * sum(f[:n-1])
err_riemannL = 0.676 - I_riemannL

I_riemannR = h * sum(f[1::])
err_riemannR = 0.676 - I_riemannR

I_mid = h * sum(((x[:n-1]+x[1:])-1)/np.exp(x[:n-1]+x[1:])/2)

I_mid = h * sum(np.sin((x[:n-1] + x[1:])/2))
err_mid = 0.676 - I_mid

print(I_riemannL)
print(err_riemannL)
print(I_riemannR)
print(err_riemannR)
print(I_mid)
print(err_mid)
```
